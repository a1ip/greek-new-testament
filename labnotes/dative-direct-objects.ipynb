# Dative Direct Objects

```python
from greeksyntax.lowfat import *

q = lowfat("nestle1904lowfat")
```

What verbs take dative "direct objects" according to the Lowfat treebanks?

c.f. [DATIVE AS DIRECT COMPLEMENT OF VERBS (Smyth)](http://www.perseus.tufts.edu/hopper/text?doc=Perseus%3Atext%3A1999.04.0007%3Apart%3D4%3Achapter%3D42%3Asection%3D96%3Asubsection%3D87)

```python
query = """
declare default collation 'http://basex.org/collation?lang=el;strength=secondary';

for $clause in //wg,
    $verb in $clause/*[@role='v'],
    $object in $clause/*[@role='o'] 
where $object[@case='dative']
   or $object/w[@head][@case='dative']
let $lemma := $verb/@lemma
group by $lemma
order by $lemma
return $lemma
"""

print(q.xquery(query))
```
